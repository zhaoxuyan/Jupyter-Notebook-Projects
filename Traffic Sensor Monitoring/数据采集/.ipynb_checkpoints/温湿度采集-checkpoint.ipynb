## 采集温湿度

```python
#!/usr/bin/env python3
# -*- coding:utf-8 -*-
import socket
import codecs
import time

sk = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
print(sk)
address = ('10.1.156.82', 8001)
sk.connect(address)
while True:    
    inp = "030300000002c5e9"
    time.sleep(3)
    sk.send(codecs.decode(inp, 'hex'))
    data = sk.recv(1024)
    print("接受到的数据：")
    print(data)

    sd = codecs.encode(data[3:5], 'hex')
    sd = int(sd, base=16) / 100
    wd = codecs.encode(data[5:7], 'hex')
    wd = int(wd, base=16) / 100
    print("处理后的数据：")
    print("湿度：%s,温度：%s" % (sd, wd))
    print("  ")
sk.close()
```

## 更优雅的版本

```python
#!/usr/bin/env python3
# -*- coding:utf-8 -*-

import socket
import codecs
import time
import struct

def gscl(gs):
    gs = gs.replace('{int}(', 'getint(data,')
    return gs


def getint(data, start, end):
    return struct.unpack('>h', data[start:end])[0]


if __name__ == "__main__":
    sk = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    sk.settimeout(3)
    try:
        cmd = b'030300000002c5e9'
        cmd = codecs.decode(cmd, 'hex')
        while True:
            sk.sendto(cmd, ('10.1.156.82', 8001))
            try:
                data, addr = sk.recvfrom(8192)
                sdgs = "{int}(3,5)"
                sdgs = gscl(sdgs)
                sd = eval(sdgs)
                wdgs = "{int}(5,7)"
                wdgs = gscl(wdgs)
                wd = eval(wdgs)
                wd = wd / 100
                sd = sd / 100
                print(wd, sd)
                time.sleep(2)
            except socket.timeout:
                print('接收数据超时！')
    finally:
        sk.close()
```
