```python
import socket
import codecs
import time
"""
client
    connect()
    recv()
    send()
    sendall()
    
"""
sk = socket.socket(socket.AF_INET,socket.SOCK_DGRAM)
print(sk)
address = ('10.1.156.82', 8001)
sk.connect(address)
# inp = input(">>>")
inp = '030300000002c5e9'
while True:
    if inp == 'exit':
        break
    sk.send(codecs.decode(inp,'hex'))
    time.sleep(1)
    result = sk.recv(1024)
    result =codecs.encode(result,'hex')   
    print(result)

sk.close()
```
