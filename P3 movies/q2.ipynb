```python
import pandas as pd
movies = pd.read_csv('split_genre.csv')
split_co = movies['production_companies'].str.split('|', expand=True)
split_co['id'] = movies['id']
merged_back = movies.merge(split_co)
```

```python
merged_back.head(3)
```

```python
melted = pd.melt(
    merged_back, id_vars=['id','original_title','release_year','popularity','budget','revenue','cast','genre'], 
    value_vars=[0, 1, 2, 3, 4], value_name='production_companies'
).drop('variable', axis = 1
).dropna()
# melted.to_csv('id_production_companies.csv', index=False)
```

```python
new_df = melted[melted['production_companies'].isin(["Universal Pictures", "Paramount Pictures"])]
new_df.to_csv('UniversalVSParamount_genre.csv', index=False)
```
